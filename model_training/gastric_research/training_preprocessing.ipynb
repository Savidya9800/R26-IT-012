```python
# ==========================================
# PREPROCESSING & DATA EXPLORATION
# ==========================================
import os
import collections
from google.colab import drive

# 1. Define Paths
base_dir = '/content/drive/MyDrive/Colab Notebooks/R26-IT-012_All_Datasets/Gastric_Module_Images/labeled-images/upper-gi-tract'

if not os.path.exists(base_dir):
    raise FileNotFoundError(f"🚨 ERROR: Cannot find the dataset folder!\nPath checked: {base_dir}")

target_classes = {'normal-z-line': 0, 'pylorus': 0, 'erythema': 1, 'esophagitis-a': 1, 'esophagitis-b-d': 1}

# 2. Scan Images
image_paths, labels = [], []
for root, dirs, files in os.walk(base_dir):
    folder_name = os.path.basename(root)
    if folder_name in target_classes:
        label = target_classes[folder_name]
        for file in files:
            if file.endswith(('.jpg', '.jpeg', '.png')):
                image_paths.append(os.path.join(root, file))
                labels.append(label)

if len(image_paths) == 0:
    raise ValueError("🚨 ERROR: No images found. Check your folder structure.")

print(f"✅ Preprocessing Success! Found {len(image_paths)} valid images.")

# 3. Calculate Weights
class_counts = collections.Counter(labels)
total_samples = len(labels)
num_classes = 2

print("\n--- Dataset Breakdown ---")
print(f"Class 0 (Normal): {class_counts[0]} images")
print(f"Class 1 (Inflammation): {class_counts[1]} images")

# We will use this exact math in the training notebook!
print("\n✅ Data is clean and ready for training. You can move to train_model.ipynb!")
```
